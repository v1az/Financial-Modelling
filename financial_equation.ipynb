```python
import pandas as pd

# Load the CSV file
file_path = '/kaggle/input/book21/Book1.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head(), data.columns
```

```python


from sklearn.linear_model import LinearRegression
import numpy as np

# Extracting the columns
years = data['year'].values.reshape(-1, 1)
market_size = data[' market size in USD bn'].values

# Creating and fitting the linear regression model
model = LinearRegression()
model.fit(years, market_size)

# Getting the coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Linear regression formula
formula = f"Market Size (in USD bn) = {slope:.2f} * Year + {intercept:.2f}"
formula
```

```python



import matplotlib.pyplot as plt

# Plotting the data points
plt.scatter(years, market_size, color='blue', label='Actual Data')

# Plotting the regression line
plt.plot(years, model.predict(years), color='red', label='Regression Line')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Market Size (in USD bn)')
plt.title('Market Size vs Year')
plt.legend()


plt.show()

```
